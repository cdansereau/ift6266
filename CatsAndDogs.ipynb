```python
%matplotlib inline
import scipy
from scipy import ndimage
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from numpy import *
```

```python

def do_sobel(im):
    im = im.astype('int32')
    dx = ndimage.sobel(im, axis=0, mode='constant')
    dy = ndimage.sobel(im, axis=1, mode='constant')
    #dx = ndimage.sobel(im, 0)  # horizontal derivative
    #dy = ndimage.sobel(im, 1)  # vertical derivative
    mag = np.hypot(dx, dy)  # magnitude
    mag *= 255.0 / np.max(mag)  # normalize (Q&D)
    return mag

def subplot_catdog(im):
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(im)
    plt.subplot(1, 2, 2)
    new_im = do_sobel(im)
    plt.imshow(np.mean(new_im,axis=2))
    

def histeq(im,nbr_bins=256):
    # http://www.janeriksolem.net/2009/06/histogram-equalization-with-python-and.html
    #get image histogram
    imhist,bins = histogram(im.flatten(),nbr_bins,normed=True)
    cdf = imhist.cumsum() #cumulative distribution function
    cdf = 255 * cdf / cdf[1] #normalize
    
    #use linear interpolation of cdf to find new pixel values
    im2 = interp(im.flatten(),bins[:-1],cdf)
    return im2.reshape(im.shape), cdf

def normalize(arr):
    """
    Linear normalization
    http://en.wikipedia.org/wiki/Normalization_%28image_processing%29
    """
    arr = arr.astype('float')
    # Do not touch the alpha channel
    for i in range(3):
        minval = arr[...,i].min()
        maxval = arr[...,i].max()
        if minval != maxval:
            arr[...,i] -= minval
            arr[...,i] *= (255.0/(maxval-minval))
    return arr

def normalized(im):
    # do an RGB normalization
    # http://akash0x53.github.io/blog/2013/04/29/RGB-Normalization/
    norm=np.zeros(im.shape,np.float32)
    norm_rgb=np.zeros(im.shape,np.uint8)

    b=im[:,:,0]
    g=im[:,:,1]
    r=im[:,:,2]

    img_sum=b+g+r

    norm_rgb[:,:,0]=b/img_sum*255.0
    norm_rgb[:,:,1]=g/img_sum*255.0
    norm_rgb[:,:,2]=r/img_sum*255.0

    #norm_rgb=cv2.convertScaleAbs(norm)
    return norm_rgb
    
```
